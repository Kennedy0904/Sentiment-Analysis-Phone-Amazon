### Explore reviews and rating of products

#### Number of reviews and mean of rating

```python
rating = cleanDF.groupby(
    ['Product']).agg(
    {
        # Find the total number of reviews
        'Reviews': 'count',
        # Find the mean rating
        'Rating': 'mean'
    }).sort_values(
        'Reviews', ascending=False)
rating.head(20)
```

#### Plot graph of total number of reviews for top 30 products

```python
# Plot number of reviews for top 30 products using seaborn
products =cleanDF["Product"].value_counts()
plt.figure(figsize=(14,10))
chart = sns.barplot(
    products[:30].index, 
    products[:30].values,
    palette='Set1'
)
chart.set_xticklabels(chart.get_xticklabels(), rotation=90, horizontalalignment='right')
chart.set_title("Total Reviews for Top 30 Products")
```

#### Number of reviews and mean of rating for each brand

```python
cleanDF.groupby(
    ['Brand']).agg(
    {
        # Find the total number of reviews
        'Reviews': 'count',
        # Find the mean rating
        'Rating': 'mean'
    }).sort_values(
        'Reviews', ascending=False
    )
```
